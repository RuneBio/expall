```python
def codon_counts(series):
    codon_string = series['true_nt_seq']
    codon_list = list(map(''.join, zip(*[iter(codon_string)]*3)))
    codon_counter = Counter(codon_list)
    my_series = pd.Series(codon_counter)
    series[my_series.index] = my_series
```

# onto my own thang w/ full data set

```python
# first... let's get the list of codons from our pals Sastry et al
featurized_data = pd.read_csv('../dataframes/DF_prest_features_1.csv', index_col=0)
codon_list = featurized_data.columns[7:71]
codon_list
```

```python
# read in full data set
data = pd.read_csv('../dataframes/DF_prest.csv', index_col=0)
```

```python
# initialize 'true_nt_seq' to mirror featurized data nomenclature
data['true_nt_seq'] = ''

# initialize codon count columns
for codon in codon_list:
    data[codon] = 0
```

```python
data
```

```python
# list of IDs for sequences with codon fragments
weird_seq_len = []
# iterate through and update rows. LOOK FOR BETTER APPROACH
# altering the object you're iterating through is supposedly
# bad practice... not sure why, but it says so in the pandas docs
for idx, series in data.iterrows():
    true_nt_seq = series['nt_seq'].replace('GACAAGCTTGCGGCCGCA', '', 1).upper()
    # check for fragmented final codon
    if len(true_nt_seq)%3:
        weird_seq_len.append(idx)
        continue
        
    # if the detagged sequence is divisible by 3, count codons
    data.loc[idx, 'true_nt_seq'] = true_nt_seq
    codons = list(map(''.join, zip(*[iter(true_nt_seq)]*3)))
    codon_counter = Counter(codons)
    my_series = pd.Series(codon_counter)
    data.loc[idx, my_series.index] = my_series
```
